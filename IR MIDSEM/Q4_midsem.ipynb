```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-0,50,1)
y = np.arange(-0,50,1)

# Goal is at (40,40) 
goal = [40,40]

#obstacle is at(25,25)
obstacle = [25,25]                                                    
obstacle1 = [35,40]
obstacle2 = [10,25]
obstacle3= [35,25]
obstacle4= [10,40]
obstacle5= [10,10]
obstacle6= [35,10]


X, Y = np.meshgrid(x,y)

delx = np.zeros_like(X)
dely = np.zeros_like(Y)

"""
    Inside the nested loop, distance from each point to the goal and ostacle is 
    calculated, Similarly angles are calculated. Then I simply used the formula give and 
    superimposed it to the Goals potential field.Also
     = 50
     = 50
    s = 15
    r = 2
"""
s = 15
r=2
for i in range(len(x)):
    for j in range(len(y)):
    
        # finding the goal distance and obstacle distance
        d_goal = np.sqrt((goal[0]-X[i][j])**2 + ((goal[1]-Y[i][j]))**2)
        d_obstacle = np.sqrt((obstacle[0]-X[i][j])**2 + (obstacle[1]-Y[i][j])**2)
        d_obstacle1 = np.sqrt((obstacle1[0]-X[i][j])**2 + (obstacle1[1]-Y[i][j])**2)
        d_obstacle2 = np.sqrt((obstacle2[0]-X[i][j])**2 + (obstacle2[1]-Y[i][j])**2)
        d_obstacle3 = np.sqrt((obstacle3[0]-X[i][j])**2 + (obstacle3[1]-Y[i][j])**2)
        d_obstacle4 = np.sqrt((obstacle4[0]-X[i][j])**2 + (obstacle4[1]-Y[i][j])**2)
        d_obstacle5 = np.sqrt((obstacle5[0]-X[i][j])**2 + (obstacle5[1]-Y[i][j])**2)
        d_obstacle6 = np.sqrt((obstacle6[0]-X[i][j])**2 + (obstacle6[1]-Y[i][j])**2)
        #print(f"{i} and {j}")

        #finding theta correspoding to the goal and obstacle 
        theta_goal= np.arctan2(goal[1] - Y[i][j], goal[0]  - X[i][j])
        theta_obstacle = np.arctan2(obstacle[1] - Y[i][j], obstacle[0]  - X[i][j])
        theta_obstacle1 = np.arctan2(obstacle1[1] - Y[i][j], obstacle1[0]  - X[i][j])
        theta_obstacle2 = np.arctan2(obstacle2[1] - Y[i][j], obstacle2[0]  - X[i][j])
        theta_obstacle3 = np.arctan2(obstacle3[1] - Y[i][j], obstacle3[0]  - X[i][j])
        theta_obstacle4 = np.arctan2(obstacle4[1] - Y[i][j], obstacle4[0]  - X[i][j])
        theta_obstacle5 = np.arctan2(obstacle5[1] - Y[i][j], obstacle5[0]  - X[i][j])
        theta_obstacle6 = np.arctan2(obstacle6[1] - Y[i][j], obstacle6[0]  - X[i][j])


        if d_obstacle < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle))  +0
        elif d_obstacle>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle)* np.cos(theta_obstacle)
            dely[i][j] = -120 * (s+r-d_obstacle)*  np.sin(theta_obstacle) 

        if d_obstacle1 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle1)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle1))  +0
        elif d_obstacle1>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle1))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle1<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle1)* np.cos(theta_obstacle1)
            dely[i][j] = -120 * (s+r-d_obstacle1)*  np.sin(theta_obstacle1) 


        if d_obstacle2 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle2)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle2))  +0
        elif d_obstacle2>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle2))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle2<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle2)* np.cos(theta_obstacle2)
            dely[i][j] = -120 * (s+r-d_obstacle2)*  np.sin(theta_obstacle2)


        if d_obstacle3 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle3)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle3))  +0
        elif d_obstacle3>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle3))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle3<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle3)* np.cos(theta_obstacle3)
            dely[i][j] = -120 * (s+r-d_obstacle3)*  np.sin(theta_obstacle3) 

        if d_obstacle4 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle4)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle4))  +0
        elif d_obstacle4>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle4))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle4<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle4)* np.cos(theta_obstacle4)
            dely[i][j] = -120 * (s+r-d_obstacle4)*  np.sin(theta_obstacle4) 

        if d_obstacle5 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle5)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle5))  +0
        elif d_obstacle5>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle5))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle5<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle5)* np.cos(theta_obstacle5)
            dely[i][j] = -120 * (s+r-d_obstacle5)*  np.sin(theta_obstacle5) 


        if d_obstacle6 < r:
            delx[i][j] = np.sign(np.cos(theta_obstacle6)) +0
            dely[i][j] = np.sign(np.cos(theta_obstacle6))  +0
        elif d_obstacle6>r+s:
            delx[i][j] = 0 +(50 * s *np.cos(theta_obstacle6))
            dely[i][j] = 0 + (50 * s *np.sin(theta_goal))
        elif d_obstacle6<r+s :
            delx[i][j] = -120 *(s+r-d_obstacle5)* np.cos(theta_obstacle6)
            dely[i][j] = -120 * (s+r-d_obstacle6)*  np.sin(theta_obstacle6) 


        if d_goal <r+s:
            if delx[i][j] != 0:
                delx[i][j]  += (50 * (d_goal-r) *np.cos(theta_goal))
                dely[i][j]  += (50 * (d_goal-r) *np.sin(theta_goal))
            else:
                delx[i][j]  = (50 * (d_goal-r) *np.cos(theta_goal))
                dely[i][j]  = (50 * (d_goal-r) *np.sin(theta_goal))

        if d_goal>r+s:
            if delx[i][j] != 0:
                delx[i][j] += 50* s *np.cos(theta_goal)
                dely[i][j] += 50* s *np.sin(theta_goal)
            else:
                delx[i][j] = 50* s *np.cos(theta_goal)
                dely[i][j] = 50* s *np.sin(theta_goal)

fig, ax = plt.subplots(figsize=(10,10))
ax.quiver(X, Y, delx, dely)
ax.add_patch(plt.Circle((20, 25), r, color='y'))
ax.add_patch(plt.Circle((35, 40), r, color='green'))
ax.add_patch(plt.Circle((10,25), r, color='blue'))
ax.add_patch(plt.Circle((35,25), r, color='red'))
ax.add_patch(plt.Circle((10, 40), r, color='brown'))
ax.add_patch(plt.Circle((10,10), r, color='black'))
ax.add_patch(plt.Circle((35, 10), r, color='orange'))

ax.annotate("Goal", xy=(33, 35), fontsize=10, ha="center")
ax.annotate("Obstacle1", xy=(20, 25), fontsize=8, ha="center")
ax.annotate("Obstacle2", xy=(10, 25), fontsize=8, ha="center")
ax.annotate("Obstacle3", xy=(35, 25), fontsize=8, ha="center")
ax.annotate("Obstacle4", xy=(10, 40), fontsize=8, ha="center")
ax.annotate("Obstacle5", xy=(9,13), fontsize=8, ha="center")
ax.annotate("Obstacle6", xy=(35,10), fontsize=8, ha="center")
ax.annotate("obstacle", xy=(35,40), fontsize=10, ha="center")

ax.set_title('Path taken by the Robot ')

plt.show() 
```
